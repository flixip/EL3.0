```python
from Service.app.utils.codeline import (
    CodeGenerator,
    CommentBlock,
    ImportBlock,
    InterfaceBlock,
    FunctionBlock,
    SingleLineBlock,
    Lang
)

# ==============================================
# 测试 1：单个代码块测试
# ==============================================
print("=" * 50)
print("测试 1：单个代码块")
print("=" * 50)

# 1.1 测试 CommentBlock（注释块）
print("\n--- 1.1 CommentBlock ---")
comment_single = CommentBlock("这是单行注释")
print("Python 单行注释：")
print(comment_single.trans("py"))
print("TypeScript 单行注释：")
print(comment_single.trans("ts"))

comment_multi = CommentBlock("这是多行注释\n支持换行", is_multiline=True)
print("Python 多行注释：")
print(comment_multi.trans("py"))
print("TypeScript 多行注释：")
print(comment_multi.trans("ts"))

# 1.2 测试 ImportBlock（导入块）
print("\n--- 1.2 ImportBlock ---")
import_py_simple = ImportBlock(module="os")
import_py_from = ImportBlock(module="pydantic", items=["BaseModel", "Field"])
import_ts_default = ImportBlock(module="axios", alias="axios")
import_ts_named = ImportBlock(module="vue", items=["ref", "reactive"])

print("Python 简单导入：")
print(import_py_simple.trans("py"))
print("Python from import：")
print(import_py_from.trans("py"))
print("TypeScript 默认导入：")
print(import_ts_default.trans("ts"))
print("TypeScript 命名导入：")
print(import_ts_named.trans("ts"))

# 1.3 测试 SingleLineBlock（单句代码块）
print("\n--- 1.3 SingleLineBlock ---")
line_py = SingleLineBlock("print('Hello World')")
line_ts = SingleLineBlock("console.log('Hello World')")

print("Python 单句：")
print(line_py.trans("py"))
print("TypeScript 单句：")
print(line_ts.trans("ts"))

# 1.4 测试 InterfaceBlock（接口/类块）
print("\n--- 1.4 InterfaceBlock ---")
fields = {
    "id": ("number", "用户ID", False),
    "name": ("string", "用户名", True),
    "age": ("number", "年龄", False)
}
interface = InterfaceBlock(name="User", fields=fields)

print("TypeScript Interface：")
print(interface.trans("ts"))
print("Python Pydantic Model：")
print(interface.trans("py"))

# 1.5 测试 FunctionBlock（函数块）
print("\n--- 1.5 FunctionBlock ---")
func_body = [
    SingleLineBlock("const res = await axios.get('/user')"),
    SingleLineBlock("return res.data")
]
func = FunctionBlock(
    name="getUser",
    params={"id": "number"},
    return_type="Promise<User>",
    body=func_body,
    is_async=True
)

print("TypeScript 异步函数：")
print(func.trans("ts"))

# ==============================================
# 测试 2：组合测试（生成完整文件）
# ==============================================
print("\n" + "=" * 50)
print("测试 2：组合生成完整 TypeScript API 文件")
print("=" * 50)

# 创建 TS 代码生成器
gen = CodeGenerator(lang="ts")

# 1. 头部
gen.add(CommentBlock("AUTO-GENERATED API"))
gen.add(CommentBlock("测试组合生成"))
gen.add(ImportBlock(module="axios", alias="axios"))
gen.add(SingleLineBlock(""))

# 2. 定义 User 接口
user_fields = {
    "id": ("number", "用户ID", False),
    "name": ("string", "用户名", False),
    "email": ("string", "邮箱", True)
}
gen.add(InterfaceBlock(name="User", fields=user_fields))
gen.add(SingleLineBlock(""))

# 3. 定义获取用户列表的响应接口
list_fields = {
    "status": ("number", "状态码", False),
    "msg": ("string", "消息", False),
    "data": ("User[]", "用户列表", False)
}
gen.add(InterfaceBlock(name="UserListResponse", fields=list_fields))
gen.add(SingleLineBlock(""))

# 4. 定义 API 函数
get_user_body = [
    SingleLineBlock("return axios.get<User>('/user', { params })")
]
gen.add(FunctionBlock(
    name="getUserApi",
    params={"params": "{ id: number }"},
    return_type="Promise<User>",
    body=get_user_body,
    is_async=True
))
gen.add(SingleLineBlock(""))

get_list_body = [
    SingleLineBlock("return axios.get<UserListResponse>('/user/list')")
]
gen.add(FunctionBlock(
    name="getUserListApi",
    params={},
    return_type="Promise<UserListResponse>",
    body=get_list_body,
    is_async=True
))

# 生成并打印完整代码
full_code = gen.generate()
print(full_code)
```
